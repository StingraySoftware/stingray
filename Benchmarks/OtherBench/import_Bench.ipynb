```python
def benchCode(benchFunc, *args):
    """
    Generalized code for benchmarking time and memory.
        
    Parameter
    ----------
    benchFunc : function
        The function to be benchmarked.
    """
    time1 = %timeit -o -q -c benchFunc(*args)
    mem1 = %memit -o -q benchFunc(*args)
    
    print(f"Best Time: {round(time1.best, 7)}, Average Time: {round(time1.average, 7)}, Worst Time: {round(time1.worst, 7)}, Std Dev: {round(time1.stdev, 7)}")
    print(f"Memory Usage: {mem1.mem_usage[0]}")
```

```python
def stingrayImport():
    import stingray

benchCode(stingrayImport)
```
